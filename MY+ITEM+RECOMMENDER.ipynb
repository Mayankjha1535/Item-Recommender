```python
####Count and Mean of different Products
```

```python
count=df.groupby('productid',as_index=False).count()
count
```

```python
#Average rating of each product
```

```python
mean=df.groupby('productid',as_index=False).mean()
mean
```

```python
#Merge count and df so that we can select items having review greater than some threshold so that it will be good for recommendation
```

```python
new_df=pd.merge(df,count,how='right',on='productid')
```

```python
#Selecting products greater than 50 reviews
```

```python
new_df=new_df.sort_values(by='userid_y',ascending=False)
df1=new_df[new_df.userid_y>=50]
```

```python
df1
```

```python
df1.shape
```

```python
#Grouping Products with their summary review
```

```python
req_df1=df1.groupby('productid')['summary_of_review_x'].apply(list)
req_df1=pd.DataFrame(req_df1)
req_df1.to_csv('req_df1.csv')
```

```python
req_df1.shape
```

```python
req_df1.head()
```

```python
req_df2=pd.read_csv('req_df1.csv')
req_df2.head()
```

```python
#Merge product with their review and average rating
```

```python
req_df3=pd.merge(req_df2,mean,how='inner',on='productid')
```

```python
req_df3.head()
```

```python
req_df3.shape
```

```python
req_df3.columns=['productid','summary_of_review','overall_rating','unixreviewtime']
req_df3.head()
```

```python
#Text cleaning in summary review
```

```python
reg=re.compile('[^a-z]+')
def clean(msg):
    msg=msg.lower()
    msg=reg.sub(' ',msg).strip()
    return msg
```

```python
req_df3['summary_of_review']=req_df3['summary_of_review'].apply(clean)
req_df3=req_df3[['productid','summary_of_review','overall_rating']]
```

```python
req_df3
```

```python
#Dealing with text data requires frequency of each word
```

```python
summarycounts=req_df3['summary_of_review']
count=CountVectorizer()
transform=count.fit_transform(summarycounts)
transform

```

```python
features=count.get_feature_names()
features
```

```python
final=pd.DataFrame(transform.A,columns=features)
final.head()
```

```python
final.to_csv('final.csv')
```

```python
#Check for datatypes of the matrix
```

```python
final.dtypes
```

```python
final=np.array(final)
```

```python
X_train,X_test=train_test_split(final,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
trainset_len=len(X_train)
trainset_len
```

```python
testset_len=len(X_test)
testset_len
```

```python
knn=NearestNeighbors().fit(X_train)
distance,indices=knn.kneighbors(X_train)
```

```python
distance
```

```python
indices
```

```python
for i in range(testset_len):
    a=knn.kneighbors([X_test[i]])
    recommended_products=a[1]
    first_recommended_product=[p[0] for p in recommended_products]
    first_recommended_product = str(first_recommended_product).strip('[]')
    first_recommended_product = int(first_recommended_product)
    second_recommended_product=[p[1] for p in recommended_products]
    second_recommended_product = str(second_recommended_product).strip('[]')
    second_recommended_product = int(second_recommended_product)
    third_recommended_product=[p[2] for p in recommended_products]
    third_recommended_product = str(third_recommended_product).strip('[]')
    third_recommended_product = int(third_recommended_product)
    fourth_recommended_product=[p[3] for p in recommended_products]
    fourth_recommended_product = str(fourth_recommended_product).strip('[]')
    fourth_recommended_product = int(fourth_recommended_product)
    fifth_recommended_product=[p[4] for p in recommended_products]
    fifth_recommended_product = str(fifth_recommended_product).strip('[]')
    fifth_recommended_product = int(fifth_recommended_product)
    print 'Based on your product reviews for {}'.format(req_df3['productid'][trainset_len+i])," average rating is {}".format(req_df3["overall_rating"][trainset_len+i])
    print 'First similar item to your product is {}'.format(req_df3['productid'][first_recommended_product])," average rating is {}".format(req_df3["overall_rating"][first_recommended_product])
    print 'Second similar item to your product is {}'.format(req_df3['productid'][second_recommended_product])," average rating is {}".format(req_df3["overall_rating"][second_recommended_product])
    print 'Third similar item to your product is {}'.format(req_df3['productid'][third_recommended_product])," average rating is {}".format(req_df3["overall_rating"][third_recommended_product])
    print 'Fourth similar item to your product is {}'.format(req_df3['productid'][fourth_recommended_product])," average rating is {}".format(req_df3["overall_rating"][fourth_recommended_product])
    print 'Fifth similar item to your product is {}'.format(req_df3['productid'][fifth_recommended_product])," average rating is {}".format(req_df3["overall_rating"][fifth_recommended_product])
    print '--------------------------------------I HOPE YOU WILL LIKE THESE ITEMS-------------------------------------------------------'
```
